```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
from collections import Counter 
```

```python
path = "../../data/results/books_info_topic_distance_born_died_"
figsize = (8, 8)
font_size = 12
colormap = 'tab10'
```

##### Years of births 

```python

for years in ["1990_1999", "2000_2009", "2010_2019"]:
    birth_years = []
    df = pd.read_excel(path + "{y}.xlsx".format(y = years))
    
    for author in df['author'].unique():
        print(author)
        year = df.loc[df['author'] == author, 'born'].values[0]
        if not(math.isnan(year)):
            birth_years.append(int(str(year)[0:3]+'0'))
    birth_years = sorted(birth_years)
    print(birth_years)
    decades = range(min(birth_years), max(birth_years)+11, 10)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 7.5))
    
    ax1.hist(birth_years, edgecolor='black', rwidth=0.8, bins=decades)#, bins=range(len(np.unique(bins)))
    ax1.set_title("Roky narození")
    print(len(np.unique(birth_years)))
    
    
    # Create a DataFrame for the pie chart
    pie_data = pd.DataFrame({'Percentage': [len(df.loc[df['authsex'] == "F: žena\">"]), len(df.loc[df['authsex'] == "M: muž\">"]), len(df.loc[df['authsex'] == "X: neuvedeno\">"])]},
                            index=['Ženy', 'Muži', 'Ostatní'])


    total_authors = len(df)
    percentage =  [len(df.loc[df['authsex'] == "F: žena\">"])/total_authors, len(df.loc[df['authsex'] == "M: muž\">"])/total_authors, len(df.loc[df['authsex'] == "X: neuvedeno\">"])/total_authors]
    lables = ['Ženy', 'Muži', 'Ostatní']
    # Plot the pie chart
    ax2.pie(percentage,autopct='%1.1f%%', labels = lables)
    #ax2 = pie_data.plot.pie(y=' ', ).get_figure() #  labels=['ženy', 'muži'],
    
    plt.sca(ax1)
    counts = Counter(birth_years) 
    _, the_count = counts.most_common(1)[0] 
    yint = range(1, the_count+2, 2)
    plt.yticks(yint)
    
    ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax2.set_title("Procentuální zastoupení genderu")
    
    
    plt.savefig('../../plots/authors_'+years)
    plt.show()
    plt.close()
   
```
